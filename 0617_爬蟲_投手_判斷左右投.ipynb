```python
from bs4 import BeautifulSoup as bs
import re
import requests
import pandas as pd
import time
import collections
import random
```

```python
data_all = pd.read_csv('./1090531_test_try_catch.csv') #2004~2020每場比賽
del data_all['Unnamed: 0'] 
#刪掉三隊
data_all = data_all[(~data_all['主隊'].isin(['中信','誠泰','米迪亞'])) & (~data_all['客隊'].isin(['中信','誠泰','米迪亞']))]
#刪掉投手與非先發
data_all = data_all[(data_all['守備位置'] != 'P')&(~data_all['守備位置'].str.startswith(' ('))]
#data_all['隊伍'].value_counts() 
```

```python
#判斷投手左右投
H_Lamigo = data_all[(data_all['隊伍'].isin(['Lamigo','樂天'])) & (data_all['主隊'].isin(['Lamigo','樂天']))] #看客投ID
# H_Lamigo['球員ID'].value_counts() #每個球員上場次數
A_Lamigo = data_all[(data_all['隊伍'].isin(['Lamigo','樂天'])) & (data_all['客隊'].isin(['Lamigo','樂天']))] #看主投ID

pitcher_ID_lst = set(H_Lamigo['客投ID'].values)
print(len(pitcher_ID_lst))
Apitcher_ID_lst = set(A_Lamigo['主投ID'].values)
print(len(Apitcher_ID_lst))
PITCHER_ID_LST = list(pitcher_ID_lst | Apitcher_ID_lst)
print(len(PITCHER_ID_LST))
```

```python
PITCHER_ID_LST
```

```python
pitcherurl = []
for id_ in PITCHER_ID_LST: #所有投手ID
    pitcherurl.append('http://www.cpbl.com.tw/players/apart.html?player_id='+ id_)
print(len(pitcherurl))
pitcherurl
```

```python
pitcher_url = pd.DataFrame(pitcherurl)
pitcher_url.columns = ['pitcher_url']
pitcher_url
```

```python
pitcher_url.to_csv('./pitcher_url.csv',index = False)
```

```python
df_total = pd.DataFrame()
num = 0
keep_craw = []
#投手左右
for url in pitcher_url: #所有投手ID
    print('現在到：'+ str(num))
    try:
        re = requests.get(url) 
        time.sleep(random.uniform(1,2))
        soup = bs(re.text, 'html.parser')
        table = soup.find_all('tr')
        lst_up_4 = table[0].text.split('\n')[1:-1]
        lst_down_4 =table[1].text.split('\n')[1:-1]
        name,no,team = soup.find('div', class_="player_info_name").text.replace('球隊:',' ').split()
        NO_team = soup.find('div','h1_side').a['href'].replace('/web/team_brief.php?&team=','')
        id_ = url.replace('http://www.cpbl.com.tw/players/apart.html?player_id=','')
        vari=[id_,NO_team,name,no,team]
        for i in lst_up_4:
            vari.append(i)
        for j in lst_down_4:
            vari.append(j)
        df=pd.DataFrame(vari).T
        df.columns=['ID','team','name', 'no','Ch_team','position','RorL','height','weight','Birthday','first','edu','ori_name']
        df_total = pd.concat([df_total,df])
    except:
        keep_craw.append(url)
        print('這網址開不了')
        print(url)
    num +=1
print(len(keep_craw))
df_total
```
