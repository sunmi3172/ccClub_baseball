```python
import requests 
import pandas as pd
from bs4 import BeautifulSoup as bs
import re
```

# 每場網址

```python
year_lst = list(range(2004,2021,1)) #2004~2020，17年
month_lst = list(range(2,11,1)) #2 ~ 10，9個月
total_game_url = []

for year in year_lst:
    for month in month_lst:
        url = 'http://www.cpbl.com.tw/schedule/index/{}-{}-01.html?&date={}-{}-01&gameno=01&sfieldsub=&sgameno=01'.format(year,month,year,month)
        re = requests.get(url) 
        soup = bs(re.text, 'html.parser')
        game = soup.find_all('div',{'class':'one_block'})
        #month_url = ['http://www.cpbl.com.tw' + game[i].a['href'] for i in range(len(game))] 不能這樣寫，延賽沒連結
        month_url = []
        for i in range(len(game)):
            try:
                month_url.append('http://www.cpbl.com.tw' + game[i].a['href'])
            except:
                pass
        print(str(year)+'年'+str(month)+'月有'+str(len(month_url)))
        if len(month_url) != 0:
            for j in month_url:
                total_game_url.append(j)
len(total_game_url)
```
