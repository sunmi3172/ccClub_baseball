```python
import requests 
import pandas as pd
from bs4 import BeautifulSoup
import bs4

# 球員個人紀錄 個人成績表 陳鏞基 三壘手
#url = "http://www.cpbl.com.tw/players/person.html?player_id=L0K0&teamno=L01"
# 球員個人紀錄 個人成績表 官大元 投手
url = "http://www.cpbl.com.tw/players/person.html?player_id=R139&teamno=E02"
# 球員個人紀錄 個人成績表 江坤宇 游擊手
#url = "http://www.cpbl.com.tw/players/person.html?player_id=004636&teamno=E02"
# 球員個人紀錄 個人成績表 周思齊 左外野手
#url = 'http://www.cpbl.com.tw/players/person.html?player_id=G072&teamno=E02'
# 球員個人紀錄 個人成績表 蘇俊璋 投手 
#url ='http://www.cpbl.com.tw/players/person.html?player_id=005558&teamno=AJL011'

response = requests.get(url)


if response.status_code != 200:
    print(print('URL發生錯誤：' + url))
    
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
df_battling = pd.DataFrame() # dict 打擊成績 
df_pitching = pd.DataFrame() # dict 投手成績
df_fielding = pd.DataFrame() # dict 守備成績
df_battler_against = pd.DataFrame() # dict 打者對戰成績
df_pitcher_against = pd.DataFrame() # dict 投手對戰成績



tables = soup.find_all("table")
for table in tables[1:]:
    trs = table.find_all("tr")
        # 抓columns
    for tr in trs:
        txs = [tx.text for tx in tr.children if tx != "\n" and not isinstance(tx, bs4.element.Comment)]       
        
        
        if txs[0] == "YEAR" and txs[-1] == "SSA":           
            
            # dict 打擊成績 
            for col in txs:
                df_battling[col] = ""
        
        elif txs[0] == "YEAR" and txs[-1] == "G/F":        
            # dict 投手成績
            for col in txs:
                df_pitching[col] = ""
            
        elif txs[0] == "YEAR" and txs[-1] == "POS":           
            # dict 守備成績
            for col in txs:
                df_fielding[col] = ""
             
        elif txs[0] == "YEAR" and txs[-1] == "FPCT":          
            # dict 歷年守備成績  
            pass
        
        elif txs[0] == "TEAM" and txs[-1] == "SSA": 
            #dict 打者對戰成績 
            for col in txs:
                df_battler_against[col] = ""
        elif txs[0] == "TEAM" and txs[-1] == "ER": 
            #dict 投手對戰成績 
            for col in txs:
                df_pitcher_against[col] = "" 

df_list  = [df_battling, df_pitching, df_fielding, df_battler_against, df_pitcher_against]
```

```python
tables = soup.find_all("table")
for table in tables[1:]:
    trs = table.find_all("tr")
    columns = [th.text for th in trs[0].find_all("th")]
    
    for df in df_list:
        if columns == list(df.columns):
            for tr in trs:
                tds = [td.text.strip() for td in tr.find_all("td")]
                if len(columns) == len(tds) and tds[0] != "TOTAL" and tds != columns :
                    df.loc[len(df)] = tds
                
    
```
